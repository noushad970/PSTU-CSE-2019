```python
import numpy as np
```

```python
# function 
# 2x+y+z=10
#3x+2y+3z=18
#x+4y+9z=16
```

```python
a=np.array([[2,1,1],[3,2,3],[1,4,9]],float)
b=np.array([[10],[18],[16]],float)
```

```python
print(a)
```

```python
print(b)
```

```python
n=len(b)
print(n)
```

```python
for i in range(0,n-1):
    m1=a[i,i]
    print("m1 us",m1)
    a[i,:]=a[i,:]/m1
    b[i]=b[i]/m1
    print(a,",")
    for j in range(i+1,n):
        m2=a[j,i]
        print("m2 is:",m2)
        a[j,:]-=m2*a[i,:]
        b[j]=b[j]-m2*b[i]
        print(a,",")
print("Matrix after gaussian elimination is:")
print(a)
```

```python
print(b)
```

```python
n=len(b)
x=np.zeros(n)
print(x)
for i in range(n-1,-1,-1):
    x[i]=b[i]
    print(x,",")
    for j in range (i+1,n):
        x[i]=x[i]-a[i,j]*x[j]
    x[i]=x[i]/a[i,i]
    print(x,",")

print("Solution Vector x: ")
print(x)
```
